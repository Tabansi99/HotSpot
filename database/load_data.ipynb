```python
# Reading all EN csv's file and merging into one data frame based on CSCE department
grade_report_paths = glob.glob('/content/drive/Shareddrives/CSCE 634/Software/Data/Grades/*EN.csv')
all_grade_report_distrib_df = pd.DataFrame() 
total_CS_records = 0
for path in grade_report_paths:
  grade_distrib_df = pd.read_csv(path)
  grade_distrib_df = grade_distrib_df.loc[grade_distrib_df['DEPT'] == 'CSCE']
  total_CS_records += len(grade_distrib_df)
  all_grade_report_distrib_df =  all_grade_report_distrib_df.append(grade_distrib_df)
print("totalCSRecords:", total_CS_records)
```

```python
# Aggregate Professor Grade Distributions across all available semester grade reports and writing to csv file
all_grades_grouped = all_grade_report_distrib_df.groupby(['INSTRUCTOR_NAME', 'COURSE_NUM']).agg({'A': 'sum', 'B': 'sum', 'C': 'sum', 'D': 'sum', 'F': 'sum', 'I': 'sum', 'S': 'sum', 'U': 'sum','Q': 'sum', 'X': 'sum'})
all_grades_grouped['A-F'] = all_grades_grouped['A'] + all_grades_grouped['B'] + all_grades_grouped['C'] + all_grades_grouped['D'] + all_grades_grouped['F']
all_grades_grouped['GPA'] = ((4 * all_grades_grouped['A']) + (3 * all_grades_grouped['B']) + (2 * all_grades_grouped['C']) + (1 * all_grades_grouped['D'])) / (all_grades_grouped['A-F'])
all_grades_grouped.to_csv("all_grades.csv")
```

```python
all_grades_csv = pd.read_csv("/content/drive/Shareddrives/CSCE 634/Software/Data/SemesterGrades/all_grades.csv")
# print(all_grades_csv.head())
professors = all_grades_csv["INSTRUCTOR_NAME"]
professor_loc = OrderedDict()
for i in range(len(professors)):
  if professors[i] not in professor_loc:
    professor_loc[professors[i]] = [i]
  else:
    # print(professors[i])
    professor_loc[professors[i]].append(i)
# print(professor_loc)

professors = list(professor_loc.keys())
for k in range(len(professors)):
  professor_name = professors[k]
  professor_row_indices = professor_loc[professor_name]
  data = {}
  for ind in professor_row_indices:
    professor_record = all_grades_csv.iloc[[ind]]
    course_code = "CSCE " + str(list(professor_record["COURSE_NUM"])[0])
    A = int(professor_record['A'])
    B = int(professor_record['B'])
    C = int(professor_record['C'])
    D = int(professor_record['D'])
    F = int(professor_record['F'])
    I = int(professor_record['I'])
    S = int(professor_record['S'])
    U = int(professor_record['U'])
    Q = int(professor_record['Q'])
    X = int(professor_record['X'])
    A_F = int(professor_record['A-F'])
    GPA = float(professor_record['GPA'])
    data[course_code] = {
        'A': A,
        'B': B,
        'C': C,
        'D': D,
        'F': F,
        'I': I,
        'S': S,
        'U': U,
        'Q': Q,
        'X': X,
        'A-F': A_F,
        'GPA': GPA
    }
  # db.collection(u'Professors').document(professor_name).set(data)
```

```python
# Adding Course Availability Information
def randN(N):
	min = pow(10, N-1)
	max = pow(10, N) - 1
	return random.randint(min, max)


fall_2021 = pd.read_csv("grd20213EN.csv")
fall_2021_csce = fall_2021.loc[fall_2021['DEPT'] == 'CSCE']
course_time_possibilities = pd.read_csv("/content/drive/Shareddrives/CSCE 634/Software/Database/Course Time Possibilities.csv")
beginTimes = course_time_possibilities["StartTime"]
endTimes = course_time_possibilities["EndTime"]
daysOfWeek = course_time_possibilities["DaysOfWeek"]
course_num_prof_list = defaultdict(set)
course_id = list(fall_2021_csce["COURSE_NUM"])
instructor_name = list(fall_2021_csce["INSTRUCTOR_NAME"])
num_cs_records = len(course_id)
for i in range(num_cs_records):
  course_num_prof_list[course_id[i]].add(instructor_name[i])

crns_used = set()
cnt = 0
for course_id in course_num_prof_list.keys():
	if course_id < 500:
		profs_per_course = list(course_num_prof_list[course_id])[0:2]
		doc_ref = db.collection(u"Semester").document("Fall 2022").collection(u"SemesterCourseOfferings").document(f"CSCE {course_id}")
		data = {}
		for prof in profs_per_course:
			crn = randN(5)
			while crn not in crns_used:
				crn = randN(5)
				crns_used.add(crn)
			randomTimeSlot = random.randint(0, len(course_time_possibilities)-1)
			bt = beginTimes[randomTimeSlot]
			et = endTimes[randomTimeSlot]
			dow = daysOfWeek[randomTimeSlot]
			crse_section = random.randint(500,540)
			max_capac = random.randint(25,100)
			cur_enrol = -1
			if 25 <= max_capac <= 45:
				cur_enrol = random.randint(5,20)
			elif 46 <= max_capac <= 66:
					cur_enrol = random.randint(21,40)
			else:
				cur_enrol = random.randint(41, 65)
			seats_rem = max_capac - cur_enrol
			data[str(crn)] = {
					"Class Timing" : {
							"BeginTime": bt,
							"DaysOfWeek": dow,
							"EndTime": et
					},
					"Course Section": crse_section,
					"Current Enrollment": cur_enrol,
					"Prof Name": prof,
					"Max Capacity": max_capac,
					"Seats Remaining": seats_rem
			}
		doc_ref.set(data)
	else:
		break
```
